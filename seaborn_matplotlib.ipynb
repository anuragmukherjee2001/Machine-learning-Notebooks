```
# We first need to import the packages that we will be using
import seaborn as sns # For plotting
import matplotlib.pyplot as plt # For showing plots

# Load in the data set
tips_data = sns.load_dataset("tips")
```

```
# Print out the first few rows of the data
tips_data.head()
```

```
tips_data
```

```
tips_data.describe()
```

```
sns.distplot(tips_data["total_bill"], kde = False).set_title("histogram of total bills")
plt.show()
```

```
sns.distplot(tips_data["tip"], kde = True).set_title("histogram of total tips")
plt.show()
```

```
sns.distplot(tips_data["size"], kde = False).set_title("histogram of sizes")
plt.show()
```

```
sns.distplot(tips_data["total_bill"], kde = False)
sns.distplot(tips_data["tip"], kde = False).set_title("histogram of tip size and total bill")
plt.show()
```

```
sns.boxplot(tips_data["total_bill"]).set_title("box plot of total bill")
plt.show()
```

```
sns.boxplot(tips_data["tip"]).set_title("box plot of total tip")
plt.show()
```

```
sns.boxplot(tips_data["total_bill"])
sns.boxplot(tips_data["tip"]).set_title("box plot of total bill and tips")
plt.show()
```

```
sns.boxplot(x = tips_data["tip"], y = tips_data["smoker"])
plt.show()
```

```
sns.boxplot(x = tips_data["tip"], y = tips_data["time"])
plt.show()
```

```
sns.boxplot(x = tips_data["tip"], y = tips_data["day"])
plt.show()
```

```
# Create a boxplot and histogram of the tips grouped by time of day
sns.boxplot(x = tips_data["tip"], y = tips_data["time"])

g = sns.FacetGrid(tips_data, row = "time")
g = g.map(plt.hist, "tip")
plt.show()
```

```
# Create a boxplot and histogram of the tips grouped by smoking status
sns.boxplot(x = tips_data["tip"], y = tips_data["smoker"])

g = sns.FacetGrid(tips_data, row = "smoker")
g = g.map(plt.hist, "tip")
plt.show()
```

```
# Create a boxplot and histogram of the tips grouped by teh day
sns.boxplot(x = tips_data["tip"], y = tips_data["day"]).set_title("box plot of tips by day")

g = sns.FacetGrid(tips_data, row = "day")
g = g.map(plt.hist, "tip")
plt.show()
```
